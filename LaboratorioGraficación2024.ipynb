```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.title("Histograma sobre la frecuencia de numero de habitaciones")
plt.xlabel("Numero de habitaciones")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1)
plt.title("Diagrama de dispersión sobre la frecuencia con que ocurre un crimen")
plt.xlabel("Valor Mediano")
plt.ylabel("Índice de crimen")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.title("Gráfico de barras de las primeras 10 ciudades por índice de crimen ")
plt.xlabel("Valor promedio en el que ocurre un crimen")
plt.ylabel("Ciudades")


plt.show()
```
